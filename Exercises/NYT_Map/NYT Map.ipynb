```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
%matplotlib inline
```

```python
cities = gpd.read_file("ne_10m_populated_places.shp")
```

```python
pd.set_option("display.max_columns", 50)
```

```python
Afghanistan = cities[cities["ADM0NAME"] == "Afghanistan"]
```

```python
Afghanistan_2 = Afghanistan[Afghanistan["SCALERANK"] > 6]
```

```python
Afghanistan_2.head()
```

```python
import geopandas

df = geopandas.GeoDataFrame(Afghanistan_2, geometry='geometry')
```

```python
df.to_file('Afghanistan.shp', driver='ESRI Shapefile')
```
