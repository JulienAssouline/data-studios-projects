```python
import pandas as pd
%matplotlib inline
import re
```

```python
df = pd.read_csv("Cost of Climate Disasters.csv")
```

```python
df.head()
```

```python
df["Cost Billions"] = df["Cost Billions"].astype(str).str.replace("CI", "").str.replace("$", "").str.replace("*", "")
```

```python
df
```

```python
df["Deaths"] = df["Deaths"].astype(str).str.replace(",", "")
```

```python
df["Hurricane"] = df["Event"].str.contains("Hurricane")
```

```python
df["Hurricane"] = df["Hurricane"].astype(str).str.replace("False", "0").str.replace("True", "1")
```

```python
df.to_csv("Cost of Climate Disasters.csv")
```
