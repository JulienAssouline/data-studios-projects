```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
years = [2000,2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012,2013, 2014, 2015, 2016, 2017, 2018]
```

```python
years_df = pd.DataFrame(years)
years_df.columns = ["Years"]
years_df["Years"] = years_df["Years"].astype(str)
```

```python
years_df
```

```python
urls = "https://www.pro-football-reference.com/years/" + years_df["Years"] + "/#all_AFC"
```

```python
pd.set_option('max_colwidth',100)
```

```python
urls
```

```python
list_nfl_data = []

for i in urls:
    response = rq.post(i)
    doc = BeautifulSoup(response.text, "html.parser")
    for n in doc.find_all("tr")[1:]:
        dictionary = {}
        Tm = n.find("a")
        if Tm is not None and len(Tm) > 0:
            dictionary["Team"] = Tm.text.strip()
        W = n.find(attrs = {"data-stat": "wins"})
        if W is not None and len(W) > 0:
            dictionary["W"] = W.text.strip()
        L = n.find(attrs = {"data-stat": "losses"})
        if L is not None and len(L) > 0:
            dictionary["L"] = L.text.strip()
        W_L = n.find(attrs = {"data-stat": "win_loss_perc"})
        if W_L is not None and len(W_L) > 0:
            dictionary["W_L"] = W_L.text.strip()
        points = n.find(attrs = {"data-stat": "points"})
        if points is not None and len(points) > 0:
            dictionary["points"] = points.text.strip()
        points_opp = n.find(attrs = {"data-stat": "points_opp"})
        if points_opp is not None and len(points_opp) > 0:
            dictionary["points_opp"] = points_opp.text.strip()
        mov = n.find(attrs = {"data-stat": "mov"})
        if mov is not None and len(mov) > 0:
            dictionary["MOV"] = mov.text.strip()
        points_diff = n.find(attrs = {"data-stat": "points_diff"})
        if points_diff is not None and len(points_diff) > 0:
            dictionary["points_diff"] = points_diff.text.strip()
        sos = n.find(attrs = {"data-stat": "sos_total"})
        if sos is not None and len(sos) > 0:
            dictionary["sos"] = sos.text.strip()
        srs = n.find(attrs = {"data-stat": "srs_total"})
        if srs is not None and len(srs) > 0:
            dictionary["srs"] = srs.text.strip()
        OSRS = n.find(attrs = {"data-stat": "srs_offense"})
        if OSRS is not None and len(OSRS) > 0:
            dictionary["OSRS"] = OSRS.text.strip()
        DSRS = n.find(attrs = {"data-stat": "srs_defense"})
        if DSRS is not None and len(DSRS) > 0:
            dictionary["DSRS"] = DSRS.text.strip()
        dictionary["Year"] = i
        list_nfl_data.append(dictionary)
        
```

```python
df = pd.DataFrame(list_nfl_data)
```

```python
df = df.dropna() 
```
