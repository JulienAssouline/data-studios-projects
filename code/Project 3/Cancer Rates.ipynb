```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams["pdf.fonttype"] = 42
```

```python
Cancer_Death_Rates = pd.read_csv("Annual Death Rates - Sheet1.csv")
Cancer_Death_Rates.head()
```

```python
ax = Cancer_Death_Rates.plot(x = "Year of Death", y = "All Races,Both Sexes", legend = False)
ax.set_title("Cancer deaths are declining dramatically in the US", loc='left', fontweight="bold")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.savefig("Cancer deaths are declining dramatically in the US.pdf")
```

```python
cancer_incidents = pd.read_csv("cancer incidents - Sheet1 (1).csv")
cancer_incidents.head()
```

```python
ax = cancer_incidents.plot(x = "Year of Diagnosis", y = "All Races,Both Sexes", legend = False)
ax.set_title("Fewer people are getting diagnosed with cancer", loc='left', fontweight="bold")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.savefig("Fewer people are getting diagnosed with cancer.pdf")
```

```python
ax = Cancer_Death_Rates.plot(x = "Year of Death", y = "All Races,Both Sexes", legend = False)
cancer_incidents.plot(x = "Year of Diagnosis", y = "All Races,Both Sexes", legend = False, ax=ax)
```
