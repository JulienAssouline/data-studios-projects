```python
url = "http://gameofthrones.wikia.com/wiki/" + episodes_2["Episodes"]

count_first = 0
first_list = []
for i in url:
    driver.get(i)
    doc_First = BeautifulSoup(driver.page_source, "html.parser")
    dictionary_first = {}
    if doc_First.find_all("h3", text = re.compile('First')):
        First_times = doc_First.find_all("h3", text = re.compile('First'))[0].find_next_sibling("").text.split("\n")
        count_first += 1
        dictionary_first["First"] = First_times
        dictionary_first["Episode"] = count_first
        print(First_times)
    elif doc_First.find_all("h4", text = re.compile('First')):
        First_times_1 = doc_First.find_all("h4", text = re.compile('First'))[0].find_next_sibling("").text.split("\n")
        count_first += 1
        dictionary_first["First"] = First_times_1
        dictionary_first["Episode"] = count_first
        print(First_times_1)
    first_list.append(dictionary_first)
```

```python
url = "http://gameofthrones.wikia.com/wiki/" + episodes_2["Episodes"]
url
count = 0
deaths_list = []
for link in url:
    response = rq.post(link)
    doc = BeautifulSoup(response.text, "html.parser")
    dictionary = {}
    if doc.find_all("h3", text = re.compile('Deaths')):
        deaths = doc.find_all("h3", text = re.compile('Deaths'))[0].find_next_sibling("").text.split("\n")
        count += 1
        dictionary["Deaths"] = deaths
        dictionary["Episode"] = count
        print(deaths, count)
    elif doc.find_all("h4", text = re.compile('Deaths')):
        deaths_2 = doc.find_all("h4", text = re.compile('Deaths'))[0].find_next_sibling("").text.split("\n")
        count += 1
        dictionary["Deaths"] = deaths_2
        dictionary["Episode"] = count
        print(deaths_2, count)
    else:
        print("NOOO RESULTS!!!", count)
    deaths_list.append(dictionary)
```
