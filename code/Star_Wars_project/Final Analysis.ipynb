```python
import pandas as pd
import re
```

```python
Star_Wars_Gender = pd.read_csv("Star_Wars_Gender.csv")
```

```python
Star_Wars_Gender_grouped = Star_Wars_Gender.groupby(["Gender", "Movie"])["Word_count"].sum().reset_index()
```

```python
Star_Wars_Gender_grouped = Star_Wars_Gender_grouped.reset_index()
```

```python
Star_Wars_Gender_grouped = Star_Wars_Gender_grouped.sort_values(by = "Movie")
```

```python
Star_Wars_Gender_grouped_sum = Star_Wars_Gender_grouped.groupby("Movie")["Word_count"].sum().reset_index()
```

```python
Star_Wars_Gender_grouped_sum.columns = ["Movie", "total"]
```

```python
Star_Wars_Gender_grouped_merged = Star_Wars_Gender_grouped.merge(Star_Wars_Gender_grouped_sum, left_on = "Movie", right_on = "Movie")
```

```python
Star_Wars_Gender_grouped_merged["proportion"] = Star_Wars_Gender_grouped_merged["Word_count"] / Star_Wars_Gender_grouped_merged["total"]* 100
```
