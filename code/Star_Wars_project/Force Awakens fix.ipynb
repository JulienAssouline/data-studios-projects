```python
Male_awakens = pd.DataFrame({"Characters": ("Finn (FN-2187)", "Han Solo", "Poe Dameron", "Kylo Ren", "General Hux", "Snoke", "Bala-Tik", "Pilot", "Unkar", "Lor San Tekka", "Captain", "Statura", "Rey / Han", "Obi Wan", "Medic", "Yoda","Unknown Man", "Emmat", "Rey / Finn", "Some Guy")})
female_awakens = pd.DataFrame({"Characters": ("Leia", "Rey", "Maz Kanata", "Phasma", "Jess", "Rey / Han", "Girl Officer", "Unknown Lady", "Woman", "Villager", "Rey / Finn")})
Robots_awakens = pd.DataFrame({"Characters": ("C-3PO", "None")})
```

```python
Robots_awakens = Robots_awakens[Robots_awakens["Characters"] != "None"]
```

```python
Male_awakens["Gender"] = "Male"
female_awakens["Gender"] = "Female"
Robots_awakens["Gender"] = "Robots"

Gender_awakens = Male_awakens.append(female_awakens)
Gender_awakens = Gender_awakens.append(Robots_awakens)
```

```python
Gender_awakens
```

```python
Corrected_Force_Awakens["Characters"] = Corrected_Force_Awakens["Characters"].str.strip()
```

```python
Corrected_Force_Awakens = Corrected_Force_Awakens.merge(Gender_awakens, left_on="Characters", right_on="Characters", how="left")
```

```python
Corrected_Force_Awakens_words = Corrected_Force_Awakens["Dialogue"].str.split(" ", 1000, expand = True)
```

```python
Corrected_Force_Awakens_words["Characters"] = Corrected_Force_Awakens["Characters"]
```

```python
Corrected_Force_Awakens_words_long = Corrected_Force_Awakens_words.melt(var_name='type', value_name='Word', id_vars=['Characters'])

Corrected_Force_Awakens_words_long["Word"].replace("", None, inplace = True)
Corrected_Force_Awakens_words_long.dropna(inplace = True)
```

```python
Corrected_Force_Awakens_words_long = Corrected_Force_Awakens_words_long.merge(Gender_awakens, left_on="Characters", right_on="Characters", how="left")
```

```python
Corrected_Force_Awakens_words_count = Corrected_Force_Awakens_words_long.groupby("Gender")["Characters"].value_counts().reset_index(name='Word_count')
```

```python
Corrected_Force_Awakens_words_count["Movie"] = "The Force Awakens"
```
