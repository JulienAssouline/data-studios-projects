```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url_clones = "http://www.imsdb.com/scripts/Star-Wars-Attack-of-the-Clones.html"
response = rq.post(url_clones)
doc_clones = BeautifulSoup(response.text, "html.parser")

```

```python
list_Clones = []
for i in doc_clones.find_all("pre"):
#     print(i.text.split("\n\n\t\t\t\t"))
    attempt = i.text.split("\n\n")
    for n in attempt:
        print(n.split(","))
        script = n.split(",")
        list_Clones.append(script)
```

```python
df_clone = pd.DataFrame(list_Clones)
```

```python
df_clone["Script"] = df_clone[0].map(str) + df_clone[1].map(str) + df_clone[2].map(str) + df_clone[3].map(str) + df_clone[4].map(str) + df_clone[5].map(str) + df_clone[6].map(str) + df_clone[7].map(str) + df_clone[8].map(str) + df_clone[9].map(str) + df_clone[10].map(str)   
```
