```python
url = "http://www.boxofficemojo.com/yearly/chart/?page=1&view=releasedate&view2=domestic&yr=2017&p=.htm"
response = rq.post(url)
doc = BeautifulSoup(response.text, "html.parser")
```

```python
movie_budget_list = []
for i in doc.find_all("tr")[9:109]:
    dictionary = {}
    dictionary["Rank"] = i.find_all("td")[0].text.strip()
    dictionary["Name"] = i.find_all("td")[1].text.strip()
    dictionary["Studio"] = i.find_all("td")[2].text.strip()
    dictionary["Total Gross"] = i.find_all("td")[3].text.strip()
    dictionary["Opening"] = i.find_all("td")[5].text.strip()
    dictionary["Open_date"] = i.find_all("td")[7].text.strip()
    dictionary["end_date"] = i.find_all("td")[8].text.strip()
    dictionary["Years"] = 2017
    print(i.find_all("td")[3].text.strip())
    movie_budget_list.append(dictionary)
```

```python
df_movie_budget = pd.DataFrame(movie_budget_list)
df_movie_budget.head()
```

```python
movie_budget_years = []

for i in urls:
    response = rq.post(i)
    doc = BeautifulSoup(response.text, "html.parser")
    for rows in doc.find_all("tr")[9:109]:
        dictionary = {}
        dictionary["Rank"] = rows.find_all("td")[0].text.strip()
        dictionary["Name"] = rows.find_all("td")[1].text.strip()
        dictionary["Studio"] = rows.find_all("td")[2].text.strip()
        dictionary["Total Gross"] = rows.find_all("td")[3].text.strip()
        dictionary["Opening"] = rows.find_all("td")[5].text.strip()
        dictionary["Years"] = 2017 
#         print(i.find_all("td")[3].text.strip())
        movie_budget_years.append(dictionary)
```
