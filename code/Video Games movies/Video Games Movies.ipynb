```python
df_video_games = pd.read_csv("Video Games turned into a movie - Sheet1.csv", error_bad_lines=False)
```

```python
df_video_games.head()
```

```python
df_video_games["Box office revenue"] = df_video_games["Box office revenue"].str.replace(",", "")
```

```python
df_video_games["Box office revenue"] = df_video_games["Box office revenue"].str.replace("$", "").str.replace("\[.*\]","")
```

```python
df_video_games["Rotten Tomatoes"] = df_video_games["Rotten Tomatoes"].str.replace("%", "")
```

```python
df_video_games.rename(index=str, columns={"Rotten Tomatoes": "Rotten Tomatoes (%)", "Metacritic": "Metacritic (%)"}, inplace = True)
```

```python
pd.set_option("display.max_rows", 350)
```

```python
df_video_games_rating = df_video_games[(df_video_games["Type"] == "Normal") | (df_video_games["Type"] == "Animation")]
```

```python
df_video_games_rating["Fresh"] = 60
```

```python
df_video_games_rating["Certified Fresh"] = 75
```

```python
df_video_games_rating = df_video_games_rating.fillna(0)
```

```python
df_video_games_rating["Rotten Tomatoes (%)"] = df_video_games_rating["Rotten Tomatoes (%)"].astype(int)
```

```python
df_video_games_rating["total"] = 100 
df_video_games_rating["rott_diff"] = df_video_games_rating["total"] - df_video_games_rating["Rotten Tomatoes (%)"]
```

```python
df_video_games_rating = df_video_games_rating.sort_values(by = "Rotten Tomatoes (%)")
```
