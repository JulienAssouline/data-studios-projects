```python
def movie_getter(row):
    try:
        driver.get("http://www.boxofficemojo.com/")
        mojo_search = driver.find_element_by_xpath('//*[@id="leftnav"]/li[2]/form/input[1]')
        mojo_search.send_keys(row["Title"])
        mojo_button = driver.find_element_by_xpath('//*[@id="leftnav"]/li[2]/form/input[2]')
        mojo_button.click()
        first_movies = driver.find_element_by_xpath('//*[@id="body"]/table[2]/tbody/tr/td/table[2]/tbody/tr[2]/td[1]/b/font/a')
        first_movies.click()
        doc = BeautifulSoup(driver.page_source, "html.parser")
        list_movies = []
        dictionary = {}
        name = doc.find_all("font", face = "Verdana")[1].text
        budget = doc.find_all("td", valign="top")[8].text
        Genre = doc.find_all("td", valign="top")[5].text
        DGross = doc.find_all("td", align="center")[2].text
        WGross = doc.find_all("td", align="right")[6].text
        return pd.Series({
            "Movies": name,
            "Budget": budget,
            "Genre": Genre,
            "DGross": DGross,
            "WGross" : WGross,
        })
    except:
        None
        

df_video_games_budget = df_video_games_rating.apply(movie_getter, axis = 1)

```
