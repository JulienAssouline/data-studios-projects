```python
url = "http://www.boxofficemojo.com/yearly/chart/?view2=worldwide&yr=2017"
response = rq.post(url)
doc = BeautifulSoup(response.text, "html.parser")
```

```python
movie_world_list = []
for i in doc.find_all("tr")[5:]:
    dictionary = {}
    print(i.find_all("td")[7].text.strip())
    dictionary["Rank"] = i.find_all("td")[0].text.strip()
    dictionary["Name"] = i.find_all("td")[1].text.strip()
    dictionary["Studio"] = i.find_all("td")[2].text.strip()
    dictionary["World_Gross"] = i.find_all("td")[3].text.strip()
    dictionary["Domestic"] = i.find_all("td")[4].text.strip()
    dictionary["percent_domestic"] = i.find_all("td")[5].text.strip()
    dictionary["percent_oversease"] = i.find_all("td")[7].text.strip()
    dictionary["Years"] = 2017
    movie_world_list.append(dictionary)
```

```python
movie_world_list_1 = []

for i in urls:
    response = rq.post(i)
    doc = BeautifulSoup(response.text, "html.parser")
    for rows in doc.find_all("tr")[5:]:
        try:
            dictionary = {}
#             print(rows.find_all("td")[1].text.strip())
            dictionary["Rank"] = rows.find_all("td")[0].text.strip()
            dictionary["Name"] = rows.find_all("td")[1].text.strip()
            dictionary["Studio"] = rows.find_all("td")[2].text.strip()
            dictionary["World_Gross"] = rows.find_all("td")[3].text.strip()
            dictionary["Domestic"] = rows.find_all("td")[4].text.strip()
            dictionary["percent_domestic"] = rows.find_all("td")[5].text.strip()
            dictionary["percent_oversease"] = rows.find_all("td")[7].text.strip()
            dictionary["Years"] = 2017
            movie_world_list_1.append(dictionary)
        except:
            print("Failed")
            pass
```

```python
df_movie_world_years = pd.read_csv("movies_world_wide_cross - movies_world_wide_cross.csv")
```

```python
df_movie_world_years["thousand"] = df_movie_world_years["World_Gross"].str.contains("k")
```

```python
df_movie_world_years["World_Gross"] = df_movie_world_years["World_Gross"].str.replace("$", "").str.replace("k", "").str.replace(",", "")
```

```python
df_movie_world_years["World_Gross"] = df_movie_world_years["World_Gross"].astype(float)
```

```python
df_movie_world_years.head()
```

```python
df_movie_world_years_Millions = df_movie_world_years.loc[df_movie_world_years["thousand"] == False]
```

```python
df_movie_world_years_Millions.head()
```

```python
df_movie_world_years_Millions["World_Gross"] = df_movie_world_years_Millions["World_Gross"] * 1000000
```

```python
df_movie_world_years_Millions.head()
```

```python
df_movie_world_years_Thousnd = df_movie_world_years.loc[df_movie_world_years["thousand"] == True]
```

```python
df_movie_world_years_Thousnd["World_Gross"] = df_movie_world_years_Thousnd["World_Gross"] * 1000
```

```python
df_movie_world_years_finished = df_movie_world_years_Millions.append(df_movie_world_years_Thousnd)
```
