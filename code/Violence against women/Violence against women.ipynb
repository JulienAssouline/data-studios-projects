```python
import pandas as pd
import re
%matplotlib inline
```

```python
df = pd.read_csv("violence against women canada.csv")
```

```python
df.head()
```

```python
df["violence_rate"] = df["violence_rate"].str.replace(",", "").str.replace("[.00]", "")
```

```python
df[df["community"] == "Yellowknife"]
```

```python
df[df["violence_rate"] == "null"]
```

```python
df.head()
```

```python
df["violence_rate"] = df["violence_rate"].str.replace("null", "0")
```

```python
df["violence_rate"] = df["violence_rate"].str.strip()
```

```python
df[df["violence_rate"] == ""]
```

```python
df["violence_rate"] = df["violence_rate"].astype(float)
```

```python
df.sort_values(by = ["year", "violence_rate"], ascending = True, inplace = True)
```

```python
df.sort_values(by = ["year", "violence_rate"], ascending = True)
```

```python
df["community_prov"] = df["community"] + "," + " " + df["province"]
```

```python
df.to_csv("violence_against_women_canada_2.csv")
```
