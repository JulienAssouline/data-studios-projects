```python
olympic_skiing_Downhill_1["minutes"] = olympic_skiing_Downhill_1["Time_end"].str.split(":", 1, expand = True)[0]
olympic_skiing_Downhill_1["delete"] = olympic_skiing_Downhill_1["Time_end"].str.split(":", 1, expand = True)[1]
```

```python
olympic_skiing_Downhill_1["seconds"] = olympic_skiing_Downhill_1["delete"].str.split(".", 1, expand = True)[0]
olympic_skiing_Downhill_1["miliseconds"] = olympic_skiing_Downhill_1["delete"].str.split(".", 1, expand = True)[1]
```

```python
olympic_skiing_Downhill_1["minutes"] = olympic_skiing_Downhill_1["minutes"].astype(int)
olympic_skiing_Downhill_1["seconds"] = olympic_skiing_Downhill_1["seconds"].astype(int)
olympic_skiing_Downhill_1["miliseconds"] = olympic_skiing_Downhill_1["miliseconds"].astype(int)
```

```python
olympic_skiing_Downhill_1["minutes"] = olympic_skiing_Downhill_1["minutes"] * 60000
olympic_skiing_Downhill_1["seconds"] = olympic_skiing_Downhill_1["seconds"] * 1000
```

```python
olympic_skiing_Downhill_1["time_miliseconds"] = olympic_skiing_Downhill_1["minutes"] + olympic_skiing_Downhill_1["seconds"] + olympic_skiing_Downhill_1["miliseconds"]
```
