```python
import pandas as pd
df_directors = pd.read_csv("Directors Birthplaces.csv", sep=",", encoding='latin-1')
df_directors.head()
```

```python
df_directors
```

```python
df_directors.replace({'birthplace_clean' : {"Yugoslavia": "Bosnia and Herzegovina", "Palestin": "State of Palestine"}}, inplace=True)
df_directors[df_directors["birthplace_clean"] == "Bosnia and Herzegovina"]
```

```python
directors_count_country = df_directors["birthplace_clean"].value_counts().reset_index(name="directors")
directors_count_country.head()
```

```python
directors_count_country.rename(index=str, columns={"index":"country"}, inplace=True)
```

```python
directors_count_country.head()
directors_count_country.shape
```

```python
world = pd.read_csv("countries.csv", sep=",", encoding='latin-1')
world.replace({'Name' : {"United States": "United States of America"}}, inplace=True)
world.head()
```

```python
Palestine = pd.DataFrame({"Name": "State of Palestine",
                         "Latitude": 31.898043, 
                         "Longitude": 35.204269,
                         "Country": "Palestine"},index=[0])
```

```python
Palestine
world = world.append(Palestine, ignore_index=True)
```

```python
directors_count_country_merged = directors_count_country.merge(world, left_on = "country", right_on = "Name")
directors_count_country_merged.shape
```

```python
directors_count_country_merged.to_csv("Directors Birthplaces Complete.csv")
```
