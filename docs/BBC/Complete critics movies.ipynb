```python
import pandas as pd
%matplotlib inline

df = pd.read_json("Complete critics movies.json")
df.head()
```

```python
df_critics_unique = df.drop_duplicates(subset='Critic name')
df_critics_unique.head()
```

```python
df_critics_count = df_critics_unique["Critic country"].value_counts().reset_index(name="movie")
df_critics_count.head()
```

```python
df_critics_count.replace({'index' : {"US": "United States", "UK": "United Kingdom", "UAE": "United Arab Emirates"}}, inplace=True)
```
