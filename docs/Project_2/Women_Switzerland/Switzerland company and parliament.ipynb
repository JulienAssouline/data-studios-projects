```python
df_parliament = pd.read_csv('parliament.csv')
```

```python
df_parliament['series'] = ""
```

```python
df_parliament.loc[df_parliament['Series Code'] == 'SG.GEN.PARL.ZS', 'series'] = 'parliament seats'
```

```python
df_parliament.loc[df_parliament['Series Code'] == 'NY.GDP.PCAP.KD', 'series'] = 'GDP per capita'
```

```python
df_parliament = df_parliament[~(df_parliament['2016 [YR2016]'] == "..")]
df_parliament
```

```python
df_parliament = df_parliament[['Country Name','series','2016 [YR2016]']].dropna()
```

```python
df_parliament['data']=df_parliament['2016 [YR2016]']
```

```python
df_parliament['country'] = df_parliament['Country Name']
df_parliament = df_parliament[['country','series','data']]
```

```python
parliament_table = df_parliament.groupby(by=('country','series'), axis=0)['data'].sum().unstack()
```

```python
parliament_table.to_csv('parliament2.csv')
```

```python
df_parliament = pd.read_csv('parliament2.csv')
df_parliament = df_parliament.dropna()
```

```python
ax = df_parliament[df_parliament['GDP per capita']>10000].plot(title='Female parliamentary representation and GDP per capita',\
                color = 'darkblue', figsize=(8,10),kind='scatter',x='GDP per capita', y='parliament seats')

x = df_parliament[df_parliament['GDP per capita']>10000]['GDP per capita']
x_swiss = df_parliament[df_parliament.country == 'Switzerland']['GDP per capita']
y = df_parliament[df_parliament['GDP per capita']>10000]['parliament seats']
y_swiss = df_parliament[df_parliament.country == 'Switzerland']['parliament seats']
fit = np.polyfit(x.astype(float), y.astype(float), deg=1)
ax.plot(x, fit[0] * x + fit[1], color='darkred', linewidth=1)
ax.plot(kind='scatter',x=x_swiss.astype(float), y=y_swiss.astype(float), color='lightblue', s=200, label='trendline')
z = np.polyfit(df_parliament.loc[:, 'GDP per capita'], df_parliament.loc[:, 'parliament seats'], 1)
p = np.poly1d(z)
df_parliament['trendline'] = p(df_parliament.loc[:, 'GDP per capita'])

df_parliament[df_parliament.country == 'Switzerland'].plot(x='GDP per capita', y='parliament seats', label='Switzerland',linewidth=8,style='rx', markersize =10, ax=ax) 
ax.set_xlim(8000,100000)
ax.set_xlabel('GDP per capita (over $10,000)')
ax.set_ylabel('proportion of parliament seats (in %)')
```
