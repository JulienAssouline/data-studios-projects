```python
Star_Wars_Trends = Star_Wars_Gender.groupby(["Years", "Gender", "Movie"])["Word_count"].sum().reset_index()
```

```python
Star_Wars_Trends = Star_Wars_Trends[Star_Wars_Trends["Movie"] != "Rogue One"]
```

```python
Star_Wars_Trends["Gender"] = Star_Wars_Trends["Gender"].replace("Droid", "Robots")
```

```python
Star_Wars_Trends_totals = Star_Wars_Trends.groupby(["Movie"])["Word_count"].sum().reset_index()
```

```python
Star_Wars_Trends_merged = Star_Wars_Trends.merge(Star_Wars_Trends_totals, left_on = "Movie", right_on = "Movie")
```

```python
Star_Wars_Trends_merged.columns = ["Years", "Gender", "Movie", "Word_count", "Total"]
```

```python
Star_Wars_Trends_merged["Percentage"]  = (Star_Wars_Trends_merged["Word_count"] / Star_Wars_Trends_merged["Total"]) * 100
```
